```python
import pandas as pd
```

```python
files = {
    'EXPGSC1': 'EX', # exports
    'GCEC1': 'G', # government expenditure
    'GDPC1': 'GDP', # gross domestic product
    'GPDIC1': 'I', # investment
    'IMPGSC1': 'IM', # imports
    'PCECC96': 'C', # consumption
    'RELACBW027SBOG': 'RELoans', # real estate loans
}

merged_df = None
for filename, var in files.items():
    print('Reading file: %s' % filename)
    df = pd.read_csv(filename+'.csv').rename(columns={filename: var, 'DATE': 'date'})
    # date format is yyyy-mm-dd, remove '-dd'
    df['date'] = df['date'].str[:-3]
    # keep only first entry for each month
    df = df.drop_duplicates(subset='date')
    if merged_df is None:
        merged_df = df
    else:
        merged_df = merged_df.merge(df, on='date')   
        
df = merged_df
df = df[df['date'].str.endswith('-01')]
df.head()
```

```python
df['EX_g'] = df['EX'].pct_change()
df['G_g'] = df['G'].pct_change()
df['GDP_g'] = df['GDP'].pct_change()
df['I_g'] = df['I'].pct_change()
df['IM_g'] = df['IM'].pct_change()
df['C_g'] = df['C'].pct_change()
df['RELoansPctGDP'] = df['RELoans'] / df['GDP']

df = df[[
    'date',
    'EX_g',
    'G_g',
    'GDP_g',
    'I_g',
    'IM_g',
    'C_g',
    'RELoansPctGDP',
]].iloc[1:]
df.to_csv('GDP-growth.csv', index=False)
```
